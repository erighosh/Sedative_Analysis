```python
bolus_only_list = []
for i, r in bolus_df.iterrows():
    if i in iv_df.index:
        continue
    else:
        bolus_only_list.append(i)

len(bolus_only_list)
```

Looping over all rows & comparing columns- if same row & col then add values

```python
ct_both = 0
ct_ivc = 0
ct_bolc = 0
ct_ivi = 0
for i, r in all_med_total.iterrows():
    # Search for same row in bolus df
    #bol_data = bolus_df.loc[i]
    #if bol_data.shape[0] > 0:
    if i in bolus_df.index:
        for col in all_med_total.columns:
            if col != 'Unnamed: 0':
                if (col in iv_df.columns) and (col in bolus_df.columns):
                    #Checking for Nan values
                    if pd.isnull(iv_df.ix[i, col]):
                        all_med_total.ix[i, col] = bolus_df.ix[i, col]
                    elif pd.isnull(bolus_df.ix[i, col]):
                        all_med_total.ix[i, col] = iv_df.ix[i, col]
                    else:
                        all_med_total.ix[i, col] = iv_df.ix[i, col] + bolus_df.ix[i, col]
                    ct_both = ct_both + 1
                elif col in iv_df.columns:
                    all_med_total.ix[i, col] = iv_df.ix[i, col]
                    ct_ivc = ct_ivc + 1
                else:
                    all_med_total.ix[i, col] = bolus_df.ix[i, col]
                    ct_bolc = ct_bolc + 1
    else:
        for col in iv_df.columns:
            if col != 'Unnamed: 0':
                all_med_total.ix[i, col] = iv_df.ix[i, col]
        ct_ivi = ct_ivi + 1

print ct_both, ct_ivc, ct_bolc, ct_ivi
```
