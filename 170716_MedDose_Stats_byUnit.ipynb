```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

Reading cohort file

```python
cohort_df = pd.read_csv('cohort2474.csv')
cohort_df.head()
```

Reading med total file

```python
cohort_med_total = pd.read_csv('170726_cohort_IV_bolus_total.csv')
cohort_med_total.head()
```

### Merging last care unit & duration of ventilation with med total file. Unnamed: 0 is the ICU stay ID

```python
cohort_unit = cohort_df[['icustay_id', 'last_careunit', 'dayHrs', 'nightHrs']].copy()
```

```python
cohort_med_hour = pd.merge(cohort_med_total, cohort_unit, left_on='Unnamed: 0', right_on= 'icustay_id', how = 'left')
cohort_med_hour.head()
```

```python
cohort_med_hour.shape
```

### Calculating average dose (mg/ hr)

```python
col_list = cohort_med_total.columns
len(col_list)
```

```python
for col in col_list:
    if 'D_' in col:
        cohort_med_hour[col] = cohort_med_hour[col]/cohort_med_hour['dayHrs']
    elif 'N_' in col:
        cohort_med_hour[col] = cohort_med_hour[col]/cohort_med_hour['nightHrs']
```

```python
cohort_med_hour.head()
```

### Grouping by care unit

```python
unit_med_count = cohort_med_hour.groupby(['last_careunit']).count()
unit_med_count
```

```python
unit_med_avg = cohort_med_hour.groupby(['last_careunit']).median()
```

```python
unit_med_avg
```

### Grouping by drug categories

```python
med_groups = pd.DataFrame(columns=['icuStay', 'D_Benzo', 'N_Benzo', 'D_Opiod', 'N_Opiod', 'D_Propofol', 'N_Propofol'])
med_groups['icuStay'] = cohort_med_hour['Unnamed: 0']
med_groups.head()
```

Checking for null values

```python
for i, r in cohort_med_hour.iterrows():
    if not pd.isnull(cohort_med_hour.ix[i, 'D_Lorazepam (Ativan)']):
        if not pd.isnull(cohort_med_hour.ix[i, 'D_Midazolam (Versed)']):
            if not pd.isnull(cohort_med_hour.ix[i, 'D_Diazepam (Valium)']):
                med_groups.ix[i,'D_Benzo'] = cohort_med_hour.ix[i,'D_Lorazepam (Ativan)'] 
                + (cohort_med_hour.ix[i, 'D_Midazolam (Versed)']/3) 
                + (cohort_med_hour.ix[i, 'D_Diazepam (Valium)']/5)
            else:
                med_groups.ix[i,'D_Benzo'] = cohort_med_hour.ix[i,'D_Lorazepam (Ativan)'] 
                + (cohort_med_hour.ix[i, 'D_Midazolam (Versed)']/3) 
        else:
            if not pd.isnull(cohort_med_hour.ix[i, 'D_Diazepam (Valium)']):
                med_groups.ix[i,'D_Benzo'] = cohort_med_hour.ix[i,'D_Lorazepam (Ativan)']
                + (cohort_med_hour.ix[i, 'D_Diazepam (Valium)']/5)
            else:
                med_groups.ix[i,'D_Benzo'] = cohort_med_hour.ix[i,'D_Lorazepam (Ativan)'] 
    else:
        if not pd.isnull(cohort_med_hour.ix[i, 'D_Midazolam (Versed)']):
            if not pd.isnull(cohort_med_hour.ix[i, 'D_Diazepam (Valium)']):
                med_groups.ix[i,'D_Benzo'] = (cohort_med_hour.ix[i, 'D_Midazolam (Versed)']/3) 
                + (cohort_med_hour.ix[i, 'D_Diazepam (Valium)']/5)
            else:
                med_groups.ix[i,'D_Benzo'] = (cohort_med_hour.ix[i, 'D_Midazolam (Versed)']/3) 
        else:
            if not pd.isnull(cohort_med_hour.ix[i, 'D_Diazepam (Valium)']):
                med_groups.ix[i,'D_Benzo'] = (cohort_med_hour.ix[i, 'D_Diazepam (Valium)']/5)
            else:
                med_groups.ix[i,'D_Benzo'] = np.NaN
```
